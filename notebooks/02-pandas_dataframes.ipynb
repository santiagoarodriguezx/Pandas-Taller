## 1. Ejemplo Básico de Creación y Manipulación de un DataFrame

```python
import pandas as pd

# Crear un DataFrame a partir de un diccionario
data = {
    'Nombre': ['Ana', 'Luis', 'Marta'],
    'Edad': [23, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

```python
# Seleccionar una columna
print(df['Nombre'])
```

```python
# Filtrar filas donde la edad es mayor de 25
filtered_df = df[df['Edad'] > 25]
print(filtered_df)
```

```python
# Añadir una nueva columna
df['Salario'] = [50000, 60000, 55000]
print(df)
```

```python
# Eliminar una columna sin usar inplace=True
df = df.drop(columns=['Ciudad'])
print(df)
```

## 2. Formas de Crear DataFrames

### 2.1 A partir de diccionarios

```python
import pandas as pd

data = {
    'Nombre': ['Ana', 'Luis', 'Marta'],
    'Edad': [23, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}
df = pd.DataFrame(data)
print(df)
```

### 2.2 A partir de listas de listas

```python
data = [
    ['Ana', 23, 'Madrid'],
    ['Luis', 35, 'Barcelona'],
    ['Marta', 29, 'Valencia']
]
df = pd.DataFrame(data, columns=['Nombre', 'Edad', 'Ciudad'])
print(df)
```

### 2.3 A partir de listas de diccionarios

```python
data = [
    {'Nombre': 'Ana', 'Edad': 23, 'Ciudad': 'Madrid'},
    {'Nombre': 'Luis', 'Edad': 35, 'Ciudad': 'Barcelona'},
    {'Nombre': 'Marta', 'Edad': 29, 'Ciudad': 'Valencia'}
]
df = pd.DataFrame(data)
print(df)
```

### 2.4 A partir de arrays de NumPy

```python
import pandas as pd
import numpy as np

data = np.array([
    ['Ana', 23, 'Madrid'],
    ['Luis', 35, 'Barcelona'],
    ['Marta', 29, 'Valencia']
])
df = pd.DataFrame(data, columns=['Nombre', 'Edad', 'Ciudad'])
print(df)
```

### 2.5 A partir de Series de Pandas

```python
nombres = pd.Series(['Ana', 'Luis', 'Marta'])
edades = pd.Series([23, 35, 29])
ciudades = pd.Series(['Madrid', 'Barcelona', 'Valencia'])

df = pd.DataFrame({
    'Nombre': nombres,
    'Edad': edades,
    'Ciudad': ciudades
})
print(df)
```

## 3. Entender el Índice en DataFrames

### 3.1 Índice numérico secuencial (por defecto)

```python
import pandas as pd

data = {'Nombre': ['Ana', 'Luis', 'Marta'], 'Edad': [23, 35, 29]}
df = pd.DataFrame(data)
print(df)
```

### 3.2 Índice basado en una columna existente

```python
df.set_index('Nombre', inplace=True)
print(df)
```

### 3.3 Índice multi-nivel (MultiIndex)

```python
import pandas as pd

arrays = [['Ana', 'Ana', 'Luis', 'Luis'], ['Math', 'Science', 'Math', 'Science']]
index = pd.MultiIndex.from_arrays(arrays, names=('Nombre', 'Asignatura'))
data = {'Notas': [90, 85, 88, 92]}
df_multi = pd.DataFrame(data, index=index)
print(df_multi)
```

### 3.4 Restablecer el índice

```python
import pandas as pd

data = {
    'Nombre': ['Ana', 'Juan', 'Pedro', 'Laura'],
    'Edad': [23, 25, 30, 28],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia']
}
df = pd.DataFrame(data)
df.set_index('Nombre', inplace=True)
print('DataFrame con indice personalizado:')
print(df)

df.reset_index(inplace=True)
print('DataFrame despues de restablecer el indice:')
print(df)
```

### 3.5 Renombrar el índice

```python
df.index.name = 'Identificador'
print(df)
```

### 3.6 Seleccionar datos mediante el índice con `loc`

```python
df.set_index('Nombre', inplace=True)
fila_ana = df.loc['Ana']
print(fila_ana)
```

### 3.7 Índice temporal (fechas)

```python
import pandas as pd

date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
df_fechas = pd.DataFrame(date_rng, columns=['Fecha'])
df_fechas['Datos'] = range(10)
df_fechas.set_index('Fecha', inplace=True)
print(df_fechas)
```

```python
# Seleccionar datos de una fecha específica
print(df_fechas.loc['2023-01-03'])

# Seleccionar datos de un rango de fechas
print(df_fechas.loc['2023-01-03':'2023-01-05'])
```

### 3.8 Índices duplicados: detectar y eliminar

```python
import pandas as pd

df_dup = pd.DataFrame({'Nombre': ['Ana', 'Luis', 'Ana'], 'Edad': [23, 35, 29]})
df_dup.set_index('Nombre', inplace=True)
print("Duplicados:", df_dup.index.duplicated())

df_dup = df_dup[~df_dup.index.duplicated(keep='first')]
print(df_dup)
```

### 3.9 Acceso a datos con MultiIndex

```python
import pandas as pd

arrays = [['Ana', 'Ana', 'Luis', 'Luis'], ['Math', 'Science', 'Math', 'Science']]
index = pd.MultiIndex.from_arrays(arrays, names=('Nombre', 'Asignatura'))
data = {'Notas': [90, 85, 88, 92]}
df_multi = pd.DataFrame(data, index=index)

# Todas las asignaturas de Ana
print(df_multi.loc['Ana'])

# Nota de Luis en Math
print(df_multi.loc[('Luis', 'Math')])
```

## 6. Acceso a Datos con `iloc` y `loc`

### 6.1 `iloc` — Acceso por posición

```python
import pandas as pd

data = {'Nombre': ['Ana', 'Luis', 'Marta'], 'Edad': [23, 35, 29]}
df = pd.DataFrame(data)

print("Primera fila (iloc[0]):")
print(df.iloc[0])
```

```python
print("Primeras dos filas (iloc[0:2]):")
print(df.iloc[0:2])
```

```python
print("Primera fila, segunda columna (iloc[0, 1]):")
print(df.iloc[0, 1])
```

```python
print("Dos primeras filas y dos primeras columnas (iloc[0:2, 0:2]):")
print(df.iloc[0:2, 0:2])
```

### 6.2 `loc` — Acceso por etiqueta

```python
df.set_index('Nombre', inplace=True)

print("Fila de 'Ana' (loc['Ana']):")
print(df.loc['Ana'])
```

```python
print("Rango 'Ana':'Luis' (loc['Ana':'Luis']):")
print(df.loc['Ana':'Luis'])
```

```python
print("Fila 'Ana', columna 'Edad' (loc['Ana', 'Edad']):")
print(df.loc['Ana', 'Edad'])

print("Rango filas 'Ana':'Luis', columna 'Edad':")
print(df.loc['Ana':'Luis', 'Edad'])
```

### 6.3 Ejemplos adicionales y filtros

```python
df.reset_index(inplace=True)

print("Filas 0:1 con columnas ['Nombre', 'Edad'] (loc):")
print(df.loc[0:1, ['Nombre', 'Edad']])
```

```python
print("Filtrar filas donde Edad > 25 (loc con condición):")
print(df.loc[df['Edad'] > 25])
```

```python
print("Selección con condición compuesta (Edad > 25 y Nombre == 'Luis'):")
print(df.loc[(df['Edad'] > 25) & (df['Nombre'] == 'Luis')])
```

---
> **Nota:** `iloc` usa índices de posición (enteros) y **excluye** el límite superior del rango. `loc` usa etiquetas e **incluye** el límite superior.
