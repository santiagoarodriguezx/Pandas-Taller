## 10.2 Aplicación de funciones: apply y map

### Ejercicio 1: Lambda para elevar al cuadrado

```python
import pandas as pd

serie = pd.Series([1, 2, 3, 4, 5])

# Aplicamos una función lambda para elevar al cuadrado
serie_cuadrado = serie.apply(lambda x: x ** 2)
print('Al cuadrado:')
print(serie_cuadrado)
```

### Ejercicio 2: map() para transformar nombres a mayúsculas

```python
import pandas as pd

nombres = pd.Series(['ana', 'luis', 'marta', 'juan'])

# Usamos map() para transformar a mayúsculas
nombres_mayus = nombres.map(str.upper)
print('Nombres en mayúsculas:')
print(nombres_mayus)
```

### Ejercicio 3: apply(np.sqrt) a una Serie numérica

### Ejercicio 4: Función personalizada de clasificación con apply()

```python
import pandas as pd

serie = pd.Series([15, 55, 82, 30, 67, 91, 44])

# Definimos la función de clasificación
def clasificar(valor):
    if valor < 40:
        return 'Bajo'
    elif valor < 70:
        return 'Medio'
    else:
        return 'Alto'

serie_clasificada = serie.apply(clasificar)
print('Clasificación:')
print(serie_clasificada)
```

### Ejercicio 5: map() con un diccionario de reemplazo

```python
import pandas as pd

serie = pd.Series(['A', 'B', 'C', 'A', 'B'])

# Diccionario de reemplazo
diccionario = {'A': 'Aprobado', 'B': 'Pendiente', 'C': 'Reprobado'}

serie_mapeada = serie.map(diccionario)
print('Serie con reemplazo:')
print(serie_mapeada)
```

## 10.6 Operaciones de alineación y combinación

### Ejercicio 1: Sumar dos Series con índices distintos

```python
import pandas as pd

s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['b', 'c', 'd'])

# Los índices no coincidentes generan NaN
suma = s1 + s2
print('Suma con índices distintos:')
print(suma)
```

### Ejercicio 2: Combinar con add() y fill_value=0

```python
import pandas as pd

s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['b', 'c', 'd'])

# fill_value=0 trata los faltantes como 0 en vez de NaN
suma_fill = s1.add(s2, fill_value=0)
print('Suma con fill_value=0:')
print(suma_fill)
```

### Ejercicio 3: Fusionar dos Series con combine_first()

```python
import pandas as pd
import numpy as np

s1 = pd.Series([1, np.nan, 3, np.nan], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

# combine_first() rellena los NaN de s1 con los valores de s2
resultado = s1.combine_first(s2)
print('Resultado de combine_first():')
print(resultado)
```

### Ejercicio 4: Índices no coincidentes y alineación

```python
import pandas as pd

s1 = pd.Series([100, 200, 300], index=['x', 'y', 'z'])
s2 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])

# Sin índices en común, todo resulta NaN
resultado = s1 + s2
print('Sin índices en común — todo queda NaN:')
print(resultado)
```

### Ejercicio 5: Operaciones aritméticas complejas

```python
import pandas as pd

s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])

# Operación: (s1 * s2 + s1) / s2
resultado = (s1 * s2 + s1) / s2
print('Resultado de (s1 * s2 + s1) / s2:')
print(resultado)
```

```python
import pandas as pd

serie = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

# Reindexamos añadiendo 'e', que quedará como NaN
serie_reind = serie.reindex(['a', 'b', 'c', 'd', 'e'])
print('Serie reindexada:')
print(serie_reind)
```

### Ejercicio 2: Reindexar en orden inverso

```python
import pandas as pd

serie = pd.Series([100, 200, 300, 400, 500], index=[1, 2, 3, 4, 5])

serie_inv = serie.reindex([5, 4, 3, 2, 1])
print('Serie en orden inverso:')
print(serie_inv)
```

### Ejercicio 3: reindex() con fill_value

```python
import pandas as pd

serie = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# Los índices nuevos se rellenan con 0
serie_fill = serie.reindex(['a', 'b', 'c', 'd', 'e'], fill_value=0)
print('Serie reindexada con fill_value=0:')
print(serie_fill)
```

### Ejercicio 4: Rellenar faltantes con el promedio

```python
import pandas as pd

serie = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

promedio = serie.mean()
serie_prom = serie.reindex(['a', 'b', 'c', 'd', 'e', 'f'], fill_value=promedio)
print(f'Promedio: {promedio}')
print('\nSerie reindexada con promedio:')
print(serie_prom)
```

### Ejercicio 5: reindex() con valores nuevos y existentes

```python
import pandas as pd

serie = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# Nuevo índice: mezcla de existentes y nuevos
nuevo_indice = ['a', 'c', 'e', 'f', 'g']
serie_mix = serie.reindex(nuevo_indice)
print('Reindexada con valores nuevos y existentes:')
print(serie_mix)
```

¿Cómo contribuye el uso de Series en Pandas a la eficiencia y comprensión de los datos en un proyecto de análisis? 

El uso de Series en Pandas permite organizar los datos en una estructura unidimensional con índices, facilitando su acceso y análisis, además ofrece funciones integradas para realizar operaciones, filtrar y transformar datos de manera rápida, lo que reduce el código y mejora la eficiencia, también sirve como base para estructuras más complejas como los DataFrames, ayudando a comprender mejor la información y tomar decisiones más claras
