##  Ejercicios de Estructura de una Gráfica

```python
# A1. Crear una Figura con plt.figure()
fig = plt.figure(figsize=(8, 6), dpi=100)
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Figura creada con plt.figure()')
plt.show()
```

```python
# A2. Crear una Figura con plt.subplots()
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Figura creada con plt.subplots()')
plt.show()
```

```python
# A3. Crear ejes con add_subplot() + título, etiquetas, leyenda y anotaciones
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111)

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label='Seno', color='blue')
ax.plot(x, np.cos(x), label='Coseno', color='red')

# Título y etiquetas
ax.set_title('Estructura completa de una gráfica', fontsize=14)
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')

# Leyenda
ax.legend()

# Anotación
ax.annotate('Máximo', xy=(np.pi/2, 1), xytext=(2, 1.3),
            arrowprops=dict(arrowstyle='->', color='green'), color='green')

# Texto
ax.text(7, 0.8, 'Zona de oscilación', fontsize=9, color='purple')

plt.grid(True)
plt.show()
```

## C. Gráficas Univariantes

```python
# C1. Histograma básico
data = rng.standard_normal(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histograma de datos univariantes')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# C2. Histograma acumulativo
data = rng.normal(0, 1, 1000)

plt.hist(data, bins=30, cumulative=True, color='green', edgecolor='black')
plt.title('Histograma Acumulativo')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulada')
plt.show()
```

```python
# C3. Histograma normalizado (density=True)
plt.hist(data, bins=30, density=True, color='purple', alpha=0.6, edgecolor='black')
plt.title('Densidad de Probabilidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

```python
# C4. Boxplot básico
data_box = [rng.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data_box, vert=True, patch_artist=True, showmeans=True)
plt.title('Diagrama de Caja y Bigotes')
plt.xlabel('Grupos')
plt.ylabel('Valores')
plt.show()
```

```python
# C5. Boxplot personalizado con colores
plt.boxplot(data_box, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            flierprops=dict(marker='o', color='red', alpha=0.5),
            medianprops=dict(color='red'))
plt.title('Diagrama de Caja y Bigotes Personalizado')
plt.xlabel('Grupos')
plt.ylabel('Valores')
plt.show()
```

```python
# C6. Violin Plot básico
plt.violinplot(data_box, showmeans=True, showmedians=True)
plt.title('Gráfico de Violín')
plt.xlabel('Grupos')
plt.ylabel('Valores')
plt.show()
```

```python
# C7. Violin Plot personalizado
parts = plt.violinplot(data_box, showmeans=False, showmedians=True)

for pc in parts['bodies']:
    pc.set_facecolor('lightblue')
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)
parts['cmedians'].set_edgecolor('red')

plt.title('Gráfico de Violín Personalizado')
plt.xlabel('Grupos')
plt.ylabel('Valores')
plt.show()
```

```python
# C8. Gráfica de línea simple (seno)
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Seno', color='blue', linewidth=2)
plt.title('Gráfica de Línea de una Función Seno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

```python
# C9. Gráficas de línea personalizadas (colores, estilos, marcadores)
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y, label='Seno', color='blue', linestyle='-', marker='o', markevery=10)
plt.title('Gráfica de línea personalizada')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

```python
# C10. Gráfica de línea comparativa (seno vs coseno)
plt.plot(x, y, label='Seno', color='blue', linestyle='-', marker='o', markevery=10)
plt.plot(x, y2, label='Coseno', color='red', linestyle='--', marker='x', markevery=10)
plt.title('Gráfica de Línea de Seno y Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

## D. Gráficas Bivariantes

```python
# D1. Scatter Plot simple
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='blue')
plt.title('Gráfico de dispersión de ejemplo')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
# D2. Scatter con tamaño (s) y color (c)
x = rng.random(50)
y = rng.random(50)
tamaño = rng.random(50) * 1000
colores = rng.random(50)

plt.scatter(x, y, s=tamaño, c=colores, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title('Gráfico de dispersión con tamaño y color')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
# D3. Scatter con mapa de color (cmap plasma)
x = rng.random(150)
y = rng.random(150)
colors = rng.random(150)
sizes = 500 * rng.random(150)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='plasma', edgecolors='w', linewidth=0.5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de dispersión con colores y tamaños')
plt.colorbar(scatter, label='Escala de colores')
plt.show()
```

```python
# D4. Scatter con línea de regresión
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

m, b = np.polyfit(x, y, 1)

plt.scatter(x, y, color='red')
plt.plot(x, m*x + b, color='blue')
plt.title('Gráfico de dispersión con línea de regresión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
# D5. Gráfica de línea comparativa de dos variables
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 1, 2, 3, 4, 5]

plt.plot(x, y1, label='Cuadrados', marker='o')
plt.plot(x, y2, label='Lineal', marker='s')
plt.title('Comparación de dos variables')
plt.xlabel('Variable X')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

```python
# D6. Línea con regresión cuadrática
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

coef = np.polyfit(x, y, 2)
polinomio = np.poly1d(coef)

plt.plot(x, y, 'o', label='Datos')
plt.plot(x, polinomio(x), label='Regresión cuadrática')
plt.title('Ajuste de regresión en gráfica de línea')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```

```python
# D7. Heatmap básico con imshow()
data = rng.random((10, 10))

plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Intensidad')
plt.title('Diagrama de calor básico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# D8. Heatmap personalizado (cmap, ticks, interpolación)
data = rng.random((10, 10))

plt.imshow(data, cmap='hot', interpolation='bicubic')
plt.colorbar(label='Intensidad')
plt.title('Diagrama de calor personalizado')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.xticks(ticks=np.arange(10), labels=[f'X{i}' for i in range(10)])
plt.yticks(ticks=np.arange(10), labels=[f'Y{i}' for i in range(10)])
plt.show()
```

```python
# D9. Heatmap de matriz de correlación
data = rng.random((100, 5))
correlacion = np.corrcoef(data, rowvar=False)

plt.imshow(correlacion, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlación')
plt.title('Diagrama de calor de matriz de correlación')
plt.xlabel('Variable')
plt.ylabel('Variable')
plt.show()
```

```python
# D10. Heatmap con anotaciones numéricas
data_small = rng.random((5, 5))

plt.imshow(data_small, cmap='YlGnBu')
plt.colorbar(label='Intensidad')
plt.title('Diagrama de calor con anotaciones')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

for i in range(data_small.shape[0]):
    for j in range(data_small.shape[1]):
        plt.text(j, i, f'{data_small[i, j]:.2f}', ha='center', va='center', color='black', fontsize=9)

plt.show()
```

## E. Gráficas Categóricas

```python
# E1. Barras apiladas
categorias = ['A', 'B', 'C', 'D']
valores1 = [3, 2, 5, 7]
valores2 = [4, 6, 2, 5]

plt.bar(categorias, valores1, label='Conjunto 1')
plt.bar(categorias, valores2, bottom=valores1, label='Conjunto 2')
plt.title('Gráfico de barras apiladas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.show()
```

```python
# E2. Barras agrupadas
ancho = 0.35
x = np.arange(len(categorias))

plt.bar(x - ancho/2, valores1, width=ancho, label='Conjunto 1')
plt.bar(x + ancho/2, valores2, width=ancho, label='Conjunto 2')
plt.title('Gráfico de barras agrupadas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.xticks(x, categorias)
plt.legend()
plt.show()
```

```python
# E3. Barras apiladas personalizadas (colores, hatch)
plt.bar(categorias, valores1, color='skyblue', label='Conjunto 1', hatch='/')
plt.bar(categorias, valores2, bottom=valores1, color='lightgreen', label='Conjunto 2', hatch='\\')
plt.title('Gráfico de barras apiladas personalizado')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.show()
```

```python
# E4. Barras con etiquetas numéricas sobre cada segmento
barras1 = plt.bar(categorias, valores1, label='Conjunto 1', color='steelblue')
barras2 = plt.bar(categorias, valores2, bottom=valores1, label='Conjunto 2', color='coral')

for barra in barras1:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2,
             barra.get_y() + altura / 2,
             f'{int(altura)}', ha='center', va='center', color='white', fontweight='bold')

for barra in barras2:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2,
             barra.get_y() + altura / 2,
             f'{int(altura)}', ha='center', va='center', color='white', fontweight='bold')

plt.title('Gráfico de barras apiladas con etiquetas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.show()
```

### 1. ¿Cuál gráfica fue más útil para interpretar datos y por qué?

El heatmap de matriz de correlación fue la gráfica más útil, en el análisis de PM2.5 permite ver rápidamente si variables como el tipo de día con pico y placa, sin restricción o fines de semana tienen alguna relación con los niveles de contaminación, con una sola visualización se identifican relaciones que no son tan evidentes en tablas

También los gráficos de violín aportan bastante, muestran cómo se distribuyen los valores de PM2.5 según el tipo de día, permiten ver si en días con pico y placa los valores están más concentrados o si hay más variación, algo que un promedio no deja tan claro

### 2. ¿Qué diferencias notaste entre usar pyplot y la API orientada a objetos?

| Característica | Interfaz `pyplot` | Interfaz OO (`fig, ax`) |
|---|---|---|
| Simplicidad | Más fácil al inicio | Un poco más extensa |
| Control | Limitado | Mucho más control |
| Subgráficos | Puede ser confuso | Más claro de manejar |
| Reutilización | Poco práctico | Más ordenado |
| Uso recomendado | Pruebas rápidas | Proyectos completos |

En la práctica pyplot funciona bien para hacer gráficas rápidas, pero cuando se empiezan a comparar varios escenarios como días con y sin pico y placa se vuelve desordenado, la API orientada a objetos es más clara porque permite controlar cada gráfico directamente y evita errores

### 3. ¿Qué tipo de gráfica consideras más adecuada para tu proyecto personal?

Como el proyecto es la medición de PM2.5 en Medellín y analizar si el pico y placa influye, la elección depende de lo que se quiera ver

Para observar cambios en el tiempo la gráfica de línea es clave, permite identificar tendencias y picos de contaminación

Para comparar días con pico y placa frente a días sin restricción las gráficas de barras o los boxplots permiten ver diferencias entre grupos de forma directa

El heatmap de correlación ayuda a identificar si existe relación entre las restricciones y los niveles de PM2.5

Los violin plots permiten ver la distribución completa de los datos, no solo el promedio, lo que es útil porque los valores pueden variar bastante entre días

En general es necesario combinar varias gráficas para tener una visión más completa y poder concluir si el pico y placa tiene un impacto real
